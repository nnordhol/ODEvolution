```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from matplotlib import ticker, cm
from ODEvolution import *
from random import choice
import seaborn as sns
```

Demonstrate usage of ODEvolution package

```python
# Generate wild type and some Genotypes with random mumax and survival fractions:
gs = [Genotype(gid='WT')]+[Genotype(gid=i+1,mumax = stats.norm(1,0.2).rvs(), survival = choice([10**-i for i in range(1,4)]), n0=0.1) for i in range(1)]
```

```python
# Initialize experiment with standard parameters:
exp = Experiment(gs)
```

```python
exp.__dict__
```

```python
# Get info about genotypes from exp:
exp.from_genos('mumax')
```

```python
exp.from_genos('survival')
```

```python
# Conduct experiment:
exp.conduct_steps(cycles=15, mutate=False, steps=25)
```

```python
plt.figure()
[plt.plot(g.ts/24, g.n[:-1]) for g in exp.genotypes]
plt.yscale('log')
plt.xlabel('Days')
plt.ylabel('CFU/ml')
```

```python
# Another 15, this time without concrete steps:
exp.conduct_steps(cycles=15, mutate=False)
```

```python
plt.figure()
[plt.plot(g.ts/24, g.n[:-1]) for g in exp.genotypes]
[plt.plot(range(1,len(g.survivors)+1), g.survivors) for g in exp.genotypes]
plt.yscale('log')
plt.xlabel('Days')
plt.ylabel('CFU/ml')
```

```python
# Change experiment conditions now:
exp.add_genotypes([Genotype(survival=0.2)])
exp.volume=1
exp.sample_volume=0.01
exp.conduct(cycles=15)
```

```python
plt.figure()
[plt.plot(g.ts/24, g.n[:-1]) for g in exp.genotypes]
[plt.plot(np.arange(1+g.emerged/24., 1+(g.emerged/24.)+len(g.survivors)), g.survivors) for g in exp.genotypes]
plt.yscale('log')
plt.xlabel('Days')
plt.ylabel('CFU/ml')
```

### Which mutants can take over population?

Also check for duration until mutant takes over population!
Repeat with poisson=False

```python
simmus=np.arange(0,1.4,0.02)
simfracs=np.arange(clonefractions[-1]/10,0.2,0.001)
wts=np.zeros((len(simmus),len(simfracs)))
evs=np.zeros((len(simmus),len(simfracs)))
wt_times = np.zeros((len(simmus),len(simfracs)))
ev_times = np.zeros((len(simmus),len(simfracs)))
rs=np.zeros((len(simmus),len(simfracs)))
ss=np.zeros((len(simmus),len(simfracs)))
for i,simmu in enumerate(simmus):
    for j,simfrac in enumerate(simfracs):
        wt = Genotype(mumax = clonemus[-1], survival = clonefractions[-1], n0=10**4, gid='WT')
        mut = Genotype(mumax = simmu, survival = simfrac, n0=1, gid='Mutant')
        exp = Experiment([wt, mut])
        exp.conduct_steps(cycles=15, stop_after_first_extinct=True)
        wt.survivors[-1]
        mut.survivors[-1]
        ratio = mut.n[-2]/(wt.n[-2]+mut.n[-2])
        wts[i,j]=wt.n[-2]
        evs[i,j]=mut.n[-2]
        wt_times[i,j] = wt.ts[-1]
        ev_times[i,j] = mut.ts[-1]
        rs[i,j]=ratio
```

```python
plt.contourf(simfracs,simmus,rs,200, locator=ticker.LogLocator())
plt.title('Fraction of mutant in the population after 15 cycles')
plt.colorbar()
plt.plot(clonefractions,clonemus,marker='o',color='black',ls='None')
plt.ylabel('Growth rate [1/h]')
plt.xlabel('Survival fraction')
plt.ylim([0.1,1.4])
```

```python
plt.contourf(simfracs,simmus, rs, 200)
plt.title('Fraction of mutant in the population after 15 cycles')
plt.colorbar()
plt.plot(clonefractions,clonemus,marker='o',color='black',ls='None', mec='black')
plt.ylabel('Growth rate [1/h]')
plt.xlabel('Survival fraction')
for i,s in enumerate(['S1','S2','S3','S4','S5','S6', 'WT']):
    plt.annotate(s,(clonefractions[i]+0.005,clonemus[i]+0.05), fontsize=10, weight='bold')
```

```python
cs=plt.contourf(simfracs,simmus,np.where(rs>0.5, ev_times, np.max(ev_times))/24,5)
plt.colorbar()
plt.contour(cs, colors='k')
plt.title('Cycles until mutant takes over population')
plt.plot(clonefractions,clonemus,marker='o',color='white',ls='None', mec='black')
plt.plot(clonefractions[-1],clonemus[-1],marker='o',color='red',ls='None', mec='white')
plt.ylabel('Growth rate [h$^{-1}$]')
plt.xlabel('Survival fraction')

for i,s in enumerate(['S1','S2','S3','S4','S5','S6', 'WT']):
    plt.annotate(s,(clonefractions[i]+0.005,clonemus[i]+0.05), fontsize=10, weight='bold', color='white')
```

#### Without Poisson sampling:

```python
simmus=np.linspace(0,1.4,50)
simfracs=np.logspace(-7,0,50, base=10)
wts=np.zeros((len(simmus),len(simfracs)))
evs=np.zeros((len(simmus),len(simfracs)))
wt_times = np.zeros((len(simmus),len(simfracs)))
ev_times = np.zeros((len(simmus),len(simfracs)))
rs=np.zeros((len(simmus),len(simfracs)))
ss=np.zeros((len(simmus),len(simfracs)))
scs = rs=np.zeros((len(simmus),len(simfracs)))
for i,simmu in enumerate(simmus):
    for j,simfrac in enumerate(simfracs):
        wt = Genotype(mumax = clonemus[-1], survival = clonefractions[-1], n0=10**4, gid='WT')
        mut = Genotype(mumax = simmu, survival = simfrac, n0 = 1, gid='Mutant')
        exp = Experiment([wt, mut], poisson=False)
        exp.conduct_steps(cycles=15, stop_after_first_extinct=True)
        wt.survivors[-1]
        mut.survivors[-1]
        #ratio = mut.n[-2]/(wt.n[-2]+mut.n[-2])
        logdiffs = np.diff(np.log(mut.survivors))-np.diff(np.log(wt.survivors))
        logdiffs = logdiffs[(logdiffs!=np.inf)&(logdiffs!=np.nan)]
        sc=np.nanmean(logdiffs)
        wts[i,j]=wt.n[-2]
        evs[i,j]=mut.n[-2]
        wt_times[i,j] = wt.ts[-1]
        ev_times[i,j] = mut.ts[-1]
        #rs[i,j]=ratio
        scs[i,j] = sc
rs=evs/(evs+wts)
```

```python
wt = Genotype(mumax = clonemus[-1], survival = clonefractions[-1], n0=10**4, gid='WT')
mut = Genotype(mumax = clonemus[2], survival = clonefractions[2], n0 = 0.1, gid='Mutant')
exp = Experiment([wt, mut], poisson=False)
exp.conduct_steps(cycles=15, stop_after_first_extinct=True)
```

### Selection coefficients:

Calculate selection coefficients based on the number of survivors. This could be done based on the ratios in stationary phase as well, it seems like there is no difference (is this key to formulate it!?)

Formula for selection coefficient _s_ of two populations mutant M and WT N with frequencies of _p_ and 1-_p_, according to Chevin, 2011:

$s= \frac{d}{dt}ln(\frac{p}{1-p})=\frac{d ln M}{dt}-\frac{d ln N}{dt}$

Proof of concept: simplest case of exponential growth, no stat. phase, then s = max_M - max_N

```python
evo = Experiment([Genotype(mumax=0.9, n0=1000),Genotype(mumax=1, n0=1000)], killing=False, poisson=False,t_span=[0,7], volume=10, sample_volume=0.01)
evo.conduct_steps(cycles=15, stop_after_first_extinct=True)
gs=evo.genotypes
```

```python
evo.add_genotypes([Genotype(mumax=1, n0=0.1)])
[plt.plot(g.ts,g.n[1:]) for g in evo.genotypes]
plt.yscale('log')
```

```python
first_occurences = []
for u in np.unique(gs[0].ts[gs[0].ts%7==0])[1:]:
    first_occurences.append(np.where(gs[0].ts==u)[0][1])
    

plt.plot(gs[1].ts[first_occurences],gs[1].n[1:][first_occurences]/gs[0].n[1:][first_occurences], marker='o')
plt.plot(gs[1].ts[first_occurences],gs[1].survivors/gs[0].survivors)
plt.yscale('log')
```

We have some numerical error:

```python
plt.plot(np.diff(np.log(gs[1].n[1:]))/np.diff(gs[1].ts)-np.diff(np.log(gs[0].n[1:]))/np.diff(gs[0].ts))
```

We see that it is (almost) correct when we adjust the plot range:

```python
plt.plot(np.diff(np.log(gs[1].n[1:]))/np.diff(gs[1].ts)-np.diff(np.log(gs[0].n[1:]))/np.diff(gs[0].ts))
plt.ylim([0,0.2])
```

### Selection coefficient with killing:

```python
evo = Experiment([Genotype(mumax=1, survival=0.0006),Genotype(mumax=0.9, survival=0.006)],poisson=False, t_span=[0,12])
evo.conduct_steps(cycles=6, steps=13)
gs=evo.genotypes
[plt.plot(g.ts,g.n[1:]) for g in evo.genotypes]
plt.yscale('log')
```

At the moment, killing takes place instantaneously, i.e. at 24 hours we have 2 values. As such, the time difference is 0 and we do not get a selection coefficient for the killing, but only for growth where the wild-type has an advantage (as seen from the negative selection coefficient):

```python
plt.plot(np.diff(np.log(gs[1].n[1:]))/np.diff(gs[1].ts)-np.diff(np.log(gs[0].n[1:]))/np.diff(gs[0].ts))
```

If we plot it without normalising for time, we see spikes whenever killing occurs:

```python
plt.plot(np.diff(np.log(gs[1].n[1:]))-np.diff(np.log(gs[0].n[1:])))
```

Using the formula, we get some compounded selection coefficient (which will likely change with stationary phase)

```python
min_len=min([len(g.survivors) for g in evo.genotypes])
plt.plot(np.diff(np.log(gs[1].survivors[:min_len]))/12-np.diff(np.log(gs[0].survivors[:min_len]))/12, marker='o')
plt.ylim([0,0.2])
```

## Killing with stationary phase:

```python
evo = Experiment([Genotype(mumax=1),Genotype(mumax=0.9, survival=0.006)],poisson=False, t_span=[0,24])
evo.conduct_steps(cycles=10, steps=25)
gs=evo.genotypes
[plt.plot(g.ts,g.n[1:]) for g in evo.genotypes]
plt.yscale('log')
```

When we look at the selection coefficient without normalising by time-units, we get the selection coefficient per cycle:

```python
min_len=min([len(g.survivors) for g in evo.genotypes])
freq_mut = gs[1].survivors[:min_len]/(gs[1].survivors[:min_len]+gs[0].survivors[:min_len])

plt.plot(np.diff(np.log(freq_mut/(1-freq_mut))), marker='o')
```

- Selection coefficients converge once cells grow until stationary phase
- If cells do not reach stationary phase, selection coefficients are stable but depend on time that cells are allowed to grow for
- Selection coefficient is independent of n0 (dynamics are not though!)
- Selection coefficient increases with cell density of mutant, as can be seen by the step between cycle 3 to 5 (which is when the mutant overtakes WT in terms of density)

```python
plt.figure()

for i in range(10,26):
    evo = Experiment([Genotype(mumax=1, survival=0.0001),Genotype(mumax=0.9, survival=0.001, n0=10)],poisson=False, t_span=[0,i])
    evo.conduct_steps(cycles=10, stop_after_first_extinct=True)
    gs=evo.genotypes
    min_len=min([len(g.survivors) for g in evo.genotypes])
    freq_mut = gs[1].survivors/(gs[1].survivors+gs[0].survivors)

    plt.plot(range(1,len(freq_mut)), np.diff(np.log(freq_mut/(1-freq_mut))), marker='o', label=f'{i} hour cycle')
    print(np.mean(np.diff(np.log(gs[1].survivors))-np.diff(np.log(gs[0].survivors))))
    
plt.ylim([-4,4])
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.xlabel('Cycle')
plt.ylabel('Selection coefficient per cycle')
```
